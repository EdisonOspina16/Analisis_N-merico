```python
def posicion_falsa(funcion, a, b, tolerancia):

    if funcion(a) * funcion(b) >= 0:
        print("El intervalo no contiene una raíz.")
        return None

    iteracion = 0
    c = b - funcion(b) * (a - b) / (funcion(a) - funcion(b))
    while abs(c) > tolerancia:
        c = b - funcion(b) * (a - b) / (funcion(a)-funcion(b))
        iteracion += 1
        if abs(funcion(c))<tolerancia:
            return c, iteracion
        if funcion(a) * funcion(c )< 0:
            b = c
        else:
            a = c
    return c, iteracion
```

```python
def biseccion(funcion,a, b, tolerancia):

    if funcion(a) * funcion(b) > 0:
        print("El método de bisección no garantiza la convergencia en este intervalo.")
        return None

    iteracion = 0
    while abs(b - a) > tolerancia:
        c = (a + b) / 2
        iteracion += 1
        if abs(funcion(c)) < tolerancia:
            return c, iteracion
        if funcion(a) * funcion(c) < 0:
            b = c
        else:
            a = c
    return c, iteracion
```

```python
import matplotlib.pyplot as plt

import numpy as np

g = 9.8
c = 15
t = 9
v = 35

funcion = lambda m: (g*m/c) * (1 - np.exp((-c/m)*t)) - v


masa, iteraciones = biseccion(funcion, 40, 80, 0.001)
masa2, iteraciones2 = posicion_falsa(funcion, 40, 80, 0.0001)
print("---BISECCION---")
print(f"la masa es: {masa}, iteraciones: {iteraciones}")
print("---POSICION FALSA---")
print(f"la masa es: {masa2}, iteraciones: {iteraciones2}")

array = np.linspace(40,70,100)
plt.plot(array,funcion(array))
plt.grid()
```
