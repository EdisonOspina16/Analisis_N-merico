```python
import numpy as np
import matplotlib.pyplot as plt
from EcuacionesDiferenciales.euler.euler import euler
```

```python
def funcionTarea(t, z):
    """"
    Â¿Quien es z?: eso depende: z =[u, w]
    """
    u = z[0]
    w = z[1]

    f_1 = w
    g = 9.8
    h = 2
    f_2 = - g * ( 3 * g * u / h - 1 / 2 )
    return np.array([f_1, f_2])
```

```python
t, w = euler(funcionTarea, 0, 10, 0.01, [0,5])

x = np.array(w)[:, 0]
x_prime = np.array(w)[:, 1]

# -- impresiones -- #

print(x)
print(x_prime)
```

```python
# Grafica

plt.plot(t, x_prime, 'or')
plt.plot(t, x, 'bo')
plt.ylim(-100, 100)
```
