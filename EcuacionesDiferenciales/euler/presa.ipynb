 El estudio de modelos matemáticos para predecir la dinámica de una población de especies competentes tiene su origen en los trabajos independientes publicados en la primera época del siglo xx por A. J. Lotka y V. Volterra \\
Considere el problema de predecir la población de dos especies, una de las cuales es un depredador, cuya población en el tiempo $t$ es $x_2(t)$, que se alimenta de la otra, que es la presa, cuya población es $x_1(t)$.

Supondremos que la presa siempre tiene un suministro de comida adecuado y que su índice de natalidad en cualquier tiempo es proporcional al número de presas vivas en ese momento; es decir, el índice de natalidad (presa) es $k_1x_1(t)$.

El índice de mortalidad de la presa depende tanto del número de presas como de depredadores vivos en ese momento. Para simplicidad, suponemos un índice de mortalidad (presa) $=k_2x_1(t)x_2(t)$. El índice de natalidad del depredador, por otro lado, depende de su suministro de comida, $x_1(t)$, así como del número de depredadores disponible para propósitos de reproducción.

Por esta razón, suponemos que el índice de natalidad (depredador) es $k_3x_1(t)x_2(t)$. El índice de mortalidad del depredador será tomado como simplemente proporcional al número de depredadores vivos en ese momento; es decir, el índice de mortalidad (depredador) $= k_4x_2(t)$.\\
Puesto que $x_1(t)$ y $x_2(t)$ representan el cambio en las poblaciones de presas y depredadores, respectivamente, en relación con el tiempo, el problema se expresa mediante el sistema de ecuaciones diferenciales no lineales
\begin{eqnarray*}
x'_1(t) = k_1x_1(t) - k_2x_1(t)x_2(t) \\
x'_2(t) = k_3x_1(t)x_2(t) - k_4x_2(t)
\end{eqnarray*}
Resuelva este sistema para $0 \leqslant t \leqslant 4$, al suponer que la población inicial de la presa es 1000 y que la del depredador es 500 y que las constantes son $k_1 = 3$, $k_2 = 0.002$, $k_3 = 0.0006$, y $k_4 = 0.5$. Bosqueje una gráfica de las soluciones para este problema, al graficar ambas poblaciones con el tiempo y describa el fenómeno físico representado

```python
import numpy as np
import matplotlib.pyplot as plt
from EcuacionesDiferenciales.euler import euler
```

```python
def competencia(t, z):
    """

    ¿Quién es z?: eso depende: z =[u, w]

    """

    x1 = z[0] # Presas - Conejitos
    x2 = z[1] # Depredadores - Zorros

    # lo que cambiamos
    k_1 = 3 # constante de natalidad de las presas - depende del número de presas
    k_2 = 0.002 # Constante de mortalidad de las presas - depende de los depredadores
    k_3 = 0.0006 # Natalidad de los depredadores - depende de las presas
    k_4 = 0.5 # Constante de mortalidad de los depredadores

    presas = k_1 * x1 - k_2 * x1 * x2
    depredador = k_3 * x1 * x2 - k_4 * x2
    return np.array([presas, depredador])
```

```python
t, poblacion = euler(competencia, 0, 20, 0.0001, [1000, 500]) # [100, 500] son las condiciones iniciales
conejos = np.array(poblacion)[:, 0]
zorros = np.array(poblacion)[:, 1]
```

```python
print(conejos)
print(zorros)
```

```python
plt.figure(figsize = (8,4), dpi = 80)
plt.subplot(121)
plt.plot(t, conejos, 'g', label = 'Conejos')
plt.plot(t, zorros, 'r', label = 'Zorros')
plt.xlabel('Tiempo')
plt.ylabel('Poblaciones')
plt.title('Competencia entre especies')
plt.legend()

plt.subplot(122)
plt.plot(conejos, zorros)
plt.xlabel('zorros')
plt.ylabel('conejos')
plt.title('Diagrama de fase')
plt.tight_layout()
```
