```python
import numpy as np
import matplotlib.pyplot as plt
from EcuacionesDiferenciales.euler import euler
```

```python
def euler(f, a, b, h, c0):
    n = int((b - a) / h)
    t = [a]
    y = [c0]
    for i in range(n):
        y.append(y[i] + h * f(t[i], y[i]))
        t.append(t[i] + h)
    return t, y

f = lambda t, y: t*np.exp(3*t)-2*y

a = 0
b = 1
h = 0.25
co = 0
tiempo, w_aprox = euler(f,a,b,h,co)
print(tiempo, w_aprox)
plt.plot(tiempo, np.array(w_aprox), '*')
plt.show()
```

```python
def funcion(t, z):
    """"
    Â¿Quien es z?: eso depende: z =[u, w]
    """
    u = z[0]
    w = z[1]
    f_1 = w
    f_2 = t * np.exp(t) - t + 2 * w - u
    return np.array([f_1, f_2])
```
